# Thresholding

## Single

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('blue.png',0)
[M,N]=(img.shape)
T=125
for i in range (M):
    for j in range(N):
        if img[i,j]<T:
            img[i,j]=0
        else:
            img[i,j]=1
plt.imshow(img)
plt.title("Single Thresholding")
```

## Dual

```python
img = cv2.imread('blue.png',0)
[M,N]=(img.shape)
T1=125
T2=200
for i in range (M):
    for j in range(N):
        if (img[i,j]>=T1 and img[i,j]<=T2):
            pass
        else:
            img[i,j] =0
        
plt.imshow(img)
plt.title("Dual Thresholding")
```

## Gray Scale


```python
img = cv2.imread('blue.png',0)
A=150
B=200
for i in range (M):
    for j in range(N):
        if (img[i,j]>=A and img[i,j]<=B):
            img[i,j] =255 #l-1
        else:
            img[i,j]=0
        
plt.imshow(img)
plt.title("GrayScale Thresholding")
```
