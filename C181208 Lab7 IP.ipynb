# Importing necessary libraries and functions

```python
from PIL import Image, ImageFilter
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
%matplotlib inline
```

# Reading necessary images

```python
img = cv2.imread('Sunflower.jpg',1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.title('Original Image')
img_data=np.array(img)
```

# Task #1: Low Pass Filtering

```python
kernel_low =np.array([[1/9,1/9,1/9,1/9,1/9],
[1/9,1/9,1/9,1/9,1/9],
[1/9,1/9,1/9,1/9,1/9],
[1/9,1/9,1/9,1/9,1/9],
[1/9,1/9,1/9,1/9,1/9]])
lowpass_5=cv2.filter2D(src=img,kernel=kernel_low,ddepth=-1) #used a mask size of 5x5
out_img=np.hstack((img_data,lowpass_5))
plt.imshow(out_img)
plt.title('Low pass filtering')
```

# Task #2: Median Filtering


```python
img_noisy1 = cv2.imread('Sunflower.jpg', 0)
 
# Obtain the number of rows and columns
# of the image
m, n = img_noisy1.shape
  
# Traverse the image. For every 3X3 area,
# find the median of the pixels and
# replace the center pixel by the median
img_new1 = np.zeros([m, n])
 
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img_noisy1[i-1, j-1],
               img_noisy1[i-1, j],
               img_noisy1[i-1, j + 1],
               img_noisy1[i, j-1],
               img_noisy1[i, j],
               img_noisy1[i, j + 1],
               img_noisy1[i + 1, j-1],
               img_noisy1[i + 1, j],
               img_noisy1[i + 1, j + 1]]
         
        temp = sorted(temp)
        img_new1[i, j]= temp[4]
 
img_new1 = img_new1.astype(np.uint8)
cv2.imwrite('new_median_filtered.jpg', img_new1)
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img)
plt.imshow(img_new1)
```

```python
img2 = cv2.imread('Sunflower.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
median=cv2.medianBlur(img2,9) #used a mask size of 9x9
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(img2),plt.title('Orginal')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Median Blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```
