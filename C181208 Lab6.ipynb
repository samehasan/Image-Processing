```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('star.png',1)
img2 = cv2.imread('round.png',1)
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img2)
plt.subplot(1,2,2)
plt.imshow(img1)
```

# TASK #1: TEMPORAL PROCESSING
1 ADDING

```python
add_img = img1+img2
plt.imshow(add_img)
plt.title('Adding')
```

```python
sub_img = img1-img2
plt.imshow(sub_img)
plt.title('Subtracting')
```

```python
mul_img=img1*img2
plt.imshow((mul_img).astype(np.uint8))
plt.title('Multiplying')
```

```python
avg_img=(img1+img2)/2
plt.imshow((avg_img).astype(np.uint8))
plt.title('Signal Averaging')
```

```python
and_img = cv2.bitwise_and(img1,img2)
plt.imshow((and_img).astype(np.uint8))
plt.title('AND')
```

```python
or_img = cv2.bitwise_or(img1,img2)
plt.imshow((or_img).astype(np.uint8))
plt.title('OR')
```

```python
xor_img = cv2.bitwise_xor(img1,img2)
plt.imshow((xor_img).astype(np.uint8))
plt.title('XOR')
```

# TASK2: GEOMETRIC PROCESSING

```python
img3 = cv2.imread('tir.png',1)
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
plt.imshow(img3)
```

```python
y = 500
x= 200
h= 1000
w= 1000
crop_img=img3[x:w,y:h]
plt.imshow(crop_img)
plt.title('Clipping')
```

```python
img_rotate_90_clockwise = cv2.rotate(img3,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotate_90_clockwise)
plt.title('rotate 90 clockwise')
```

```python
img_flip_ud = cv2.flip(img3,0)
plt.imshow(img_flip_ud)
plt.title('Flipping on horizontal axis')
```

```python
height, width = img3.shape[:2]
zoom_img = cv2.resize(img3,(10*width, 10*height), interpolation = 
cv2.INTER_CUBIC)
plt.imshow(zoom_img)
plt.title('Zooming')
```
