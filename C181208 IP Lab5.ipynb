```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('blue.png',0)
```

```python
plt.imshow(img)
```

# Processing

```python
a=np.zeros((256,),dtype=np.float16)
b=np.zeros((256,),dtype=np.float16)
height,width=img.shape
```

```python
for i in range(width):
    for j in range(height):
        g = img[j,i]
        a[g] = a[g]+1

print(a)
plt.plot(a)
```

# Equalization

```python
tmp = 1.0/(height*width)
b = np.zeros((256,),dtype=np.float16)

for i in range(256):
    for j in range(i+1):
        b[i] += a[j] * tmp;
    b[i] = round(b[i] * 255);
```

```python
b=b.astype(np.uint8)

print(b)
plt.plot(b)
```

```python
for i in range(width):
    for j in range(height):
        g = img[j,i]
        img[j,i]= b[g]
```

```python
imge = cv2.imread('blue.png',0)
cv2.imshow('image',imge)
plt.subplot(1,2,1)
plt.imshow(imge)
cv2.imshow('image',img)
plt.subplot(1,2,2)
plt.imshow(img)
```
